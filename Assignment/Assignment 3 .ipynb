```
# a) Create HR_Employee Pysaprk dataFrame
from pyspark.sql.functions import *
hr_employee_df = spark.read.csv('dbfs:/FileStore/tables/HR_Employee.csv',header=True,inferSchema=True)
hr_employee_df.createOrReplaceTempView("hremployeetable")
hr_employee_df.count()
```

```
# b) Show printSchema & show 10 rows of Dataset
hr_employee_df.printSchema()
hr_employee_df.show(10)
```

```
# c) Most employee is working in which department
spark.sql("select Department,count(*) as EmployeeCount from hremployeetable group by Department order by EmployeeCount DESC").show(truncate=False)
# In Research & Development Department
```

```
# d)Highest number of job roles
spark.sql("select JobRole,count(*) as JobRoleCount from hremployeetable group by JobRole order by JobRoleCount Desc").show(truncate=False)
#In Sales Executive Role
```

```
# e) Which gender have higher strength as workforce?
spark.sql("select Gender,count(*) as GenderCount from hremployeetable group by Gender order by GenderCount Desc").show(truncate=False)
#Male
```

```
# f) Compare the marital status of employee and find the most frequent status.
spark.sql("select MaritalStatus,count(*) as MaritalCount from hremployeetable group by MaritalStatus order by MaritalCount Desc").show(truncate=False)
#Married has most count.
```

```
# g) Mostly hired employee have qualification
spark.sql("select Education,count(*) as QualificationCount from hremployeetable group by Education order by QualificationCount Desc").show(truncate=False)
#Bachelor Qualification are hired mostly.
```

```
# h) Find the count of employee from which education fields 
spark.sql("select EducationField,count(*) as EduCount from hremployeetable group by EducationField order by EduCount Desc").show()
```

```
# i) What is the job satisfaction level of employee?
spark.sql("select JobSatisfaction,count(*) as SatisfactionCount from hremployeetable group by JobSatisfaction order by SatisfactionCount Desc").show()
```

```
# j) Does most of employee do overtime: Yes or No?
spark.sql("select OverTime,count(*) as OvertimeCount from hremployeetable group by OverTime order by OvertimeCount Desc").show()
#No
```

```
# k) Find Min & Max Salaried employees.
spark.sql("select EmployeeId, Income from hremployeetable where Income = (select min(Income) from hremployeetable)").show()
spark.sql("select EmployeeId, Income from hremployeetable where Income = (select max(Income) from hremployeetable)").show()
```

```
# l) Does most of the employee do business travel? Find no.of the employees counts for each category
spark.sql("select JobRole,count(*) as TravelFrequency from hremployeetable where BusinessTravel != 'Non-Travel' group by JobRole order by TravelFrequency Desc").show(truncate=False)
```

```
# m) Find the AVG salary of graduate employee.
spark.sql("select Education,avg(Income) as AvgIncome from hremployeetable where Education = 'Bachelor' or Education = 'College' group by Education order by AvgIncome Desc").show(truncate=False)
```

```
# n) Find the employee qualification receiving salary lower than equal to avg. salary of all employee. 
spark.sql("select EmployeeID,Education,Income from hremployeetable where Income <= (select avg(Income) from hremployeetable)").show()

```
